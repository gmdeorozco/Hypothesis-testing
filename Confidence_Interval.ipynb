```python
import numpy as np
import math
```

Helper functions for Mean Calculations

```python
def get_std_error(std, n):
  std_error = std/np.sqrt(n)
  print(f'Standard error for Std: {std} and n: {n} = {std_error}')
  return std_error
```

```python
def get_margin_of_error(cv, std_error):
  m_e = cv*std_error
  print(f'Margin of error for Critical value: {cv}, Standard error: {std_error} = {m_e}')
  return m_e
```

```python
def get_confidence_interval(sample_mean, margin_of_error):
  up = sample_mean + margin_of_error
  down = sample_mean - margin_of_error
  print(f'Confidence Interval for Sample mean = {sample_mean}, Margin or error = {margin_of_error} is ( {down} , {up} )')
  return down, up
```

```python
def get_sample_size(std, critical_value, margin_of_error):
  n = math.ceil((critical_value*std/margin_of_error)**2)
  print(f'The Sample size for a Margin of error: {margin_of_error}, Populatin std: {std}, Critical value: {critical_value} = {n}')
  return n
```

Helper functions for Proportions Calculations

```python
def get_proportions_std_error(p_hat, n):
  proportions_std_error = np.sqrt(p_hat*(1-p_hat)/n)
  print(f'Proportions standard error for P Hat: {p_hat} and n: {n} = {proportions_std_error}')
  return proportions_std_error
```

```python
def get_proportions_confidence_interval(p_hat, margin_of_error):
  up = p_hat + margin_of_error
  down = p_hat - margin_of_error
  print(f'Confidence Interval for P Hat = {p_hat}, Margin or error = {margin_of_error} is ( {down} , {up} )')
  return down, up
```

#Known Standard Deviation

##1. We have a population N = 6000 adults. We peek a Sample of 49 people and want to know the Confidence Interval of 95% for the average height of adults. The sample average is 170 cm. Std of Population height is 25

```python
# We peek a sample of 49 people, the mean
n = 49
sample_ = 170 # cm is the sample mean
population_ = 25 # cm
critical_value = 1.96

standard_error = get_std_error(population_, n)
margin_of_error = get_margin_of_error(cv=critical_value, std_error=standard_error)
confidence_interval = get_confidence_interval(sample_, margin_of_error)
```
